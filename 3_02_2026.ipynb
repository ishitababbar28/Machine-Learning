```python
import numpy as np
```

```python
x = np.array([1,3,4,7,8,9,10,15])
```

```python
y = np.array([10,20,30,40,50,60,70,80])
```

```python
def Pearson_correlation(X,Y):
    if len(X)==len(Y):
        Sum_xy = sum((X-X.mean())*(Y-Y.mean()))
        Sum_x_squared = sum((X-X.mean())**2)
        Sum_y_squared = sum((Y-Y.mean())**2)
        corr = Sum_xy / np.sqrt(Sum_x_squared * Sum_y_squared)
    return corr

print(Pearson_correlation(x,y))
print(Pearson_correlation(x,x))
```

```python
x4 = np.array([1, 2, 3, 4, 5])
```

```python
def Pearson_correlation(X, Y):
    sum_xy = np.sum((X - X.mean()) * (Y - Y.mean()))
    sum_x2 = np.sum((X - X.mean())**2)
    sum_y2 = np.sum((Y - Y.mean())**2)
    return sum_xy / np.sqrt(sum_x2 * sum_y2)

# Output
print(Pearson_correlation(x4, y4))
```

```python
import pandas as pd

df = pd.read_csv("insurance2.csv")
df.head()
```
